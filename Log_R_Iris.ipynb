```python
#from sklearn import datasets

from sklearn.datasets import load_iris #
iris = datasets.load_iris()
iris
```

```python
3# efsel in from outputx
x = iris.data
x
```

```python

y = iris.target
y
```

```python
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(x, y, test_size=0.2, random_state=42) # spliting data set
```

```python
#model1 using LogisticRegression
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression() # implement LogisticRegression to the model
model1.fit(X_train, y_train) # fitting values (training)
predictions = model1.predict(X_test) # predecting target values of data set after training
score = model1.score(X_test, y_test) # fining score
print(score)
```

```python
print(metrics.classification_report(y_test, predictions)) # FINDING TARGET VARUIATION BEFORE TRAINIG AND AFTER TRAINING
```
