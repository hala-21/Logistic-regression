```python
#Logistic Regression
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split #
from sklearn.preprocessing import MinMaxScaler #
from sklearn.linear_model import LogisticRegression  #
from sklearn import metrics  #
import matplotlib.pyplot as plt #
```

```python
#from sklearn import datasets
from sklearn.datasets import load_digits
digits = load_digits()
print(digits.data.shape)
```

```python
# efsel in from outputx
x = digits.data
x
```

```python

y = digits.target
y
```

```python
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train = train_test_split(x, y, test_size=0.2, random_state=42) # spliting data set
```

```python
#model1 using LogisticRegression
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression() # implement LogisticRegression to the model
model1.fit(X_train, y_train) # fitting values (training)
predictions = model1.predict(X_test) # predecting target values of data set after training
score = model1.score(X_test, y_test) # fining score
print(score)
```

```python
print(metrics.classification_report(y_test, predictions)) # FINDING TARGET VARUIATION BEFORE TRAINIG AND AFTER TRAINING
```

```python
#TO PLOT THE classification_report WE CALL
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm=confusion_matrix(y_test, predictions, labels=model1.classes_)
cm
```

```python
#finding  recall and persisiom in one value
ConfusionMatrixDisplay.from_predictions(y_test, predictions)
```

```python
#logistivc regression with (penality term =1 give lasso )
from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression(penalty='l1', solver='liblinear', C=10) # where c = 1 / alpha (c=10 then the alpha = 0.1)
#fit
model2=model1.fit(X_train, y_train)
#predict
predictions = model2.predict(X_test)
#score
score = model2.score(X_test, y_test)
print(score)
```

```python
print(metrics.classification_report(y_test, predictions)) # FINDING TARGET VARUIATION BEFORE TRAINIG AND AFTER TRAINING
```

```python
#finding  recall and persisiom in one value
ConfusionMatrixDisplay.from_predictions(y_test, predictions)
```

```python
# repeat the last 3 steps for ridge
#logistivc regression with (penality term =l2 give rige )
from sklearn.linear_model import LogisticRegression
model3 = LogisticRegression(penalty='l2', solver='liblinear', C=10) # where c = 1 / alpha (c=10 then the alpha = 0.1)
#fit
model3=model1.fit(X_train, y_train)
#predict
predictions = model3.predict(X_test)
#score
score = model3.score(X_test, y_test)
print(score)
```

```python
print(metrics.classification_report(y_test, predictions)) # FINDING TARGET VARUIATION BEFORE TRAINIG AND AFTER TRAINING
```

```python
#finding  recall and persisiom in one value
ConfusionMatrixDisplay.from_predictions(y_test, predictions)
```
